# Подготовка данных для машинного обучения

Перед обучением модели на данных важно принять во внимание несколько ключевых аспектов:

1️⃣ Качество данных: Чем более точные данные вы используете, тем лучшие результаты можно ожидать.

2️⃣ Предобработка данных: Иногда требуется выполнить предварительную обработку данных, такую как удаление выбросов, масштабирование или нормализация признаков. Это может помочь улучшить процесс обучения и результаты модели.

3️⃣ Выбор признаков: Определите, какие признаки данных будут использоваться для обучения модели. Релевантные и информативные признаки помогут сети принимать правильные решения.

4️⃣ Разделение на тренировочную и тестовую выборки: Важно разделить данные на две непересекающиеся группы — тренировочную и тестовую выборки. Тренировочная выборка используется для обучения модели, а тестовая — для оценки ее обобщающей способности.

5️⃣ Управление переобучением: Переобучение может возникнуть, когда модель "запоминает" тренировочные данные и не может хорошо обобщать на новые данные. Используйте методы регуляризации, такие как dropout или L1/L2 регуляризация, чтобы справиться с этой проблемой.

### Поиск неявных дубликатов

Как упоминалось ранее, для «gender» 49 различных значений, и было подозрение, что некоторые из этих значений не следует рассматривать как разные категории. В конечном итоге для простоты мы разделим данные на 3 категории: мужчина, женщина и другие (сюда вошли те категории, которые можно однозначно исключить из предыдущих двух, для примера - трансгендер).

```python
male_terms = ["male", "m", "mal", "msle", "malr", "mail", "make", "cis male", "man", "maile", "male (cis)", "cis man"]
female_terms = ["female", "f", "woman", "femake", "femaile", "femake", "cis female", "cis-female/femme", "female (cis)", "femail", "cis woman"]

def clean_gender(response):
    if response.lower().rstrip() in male_terms:
        return "Male"
    elif response.lower().rstrip() in female_terms:
        return "Female"
    else:
        return "Other"

features['Gender'] = features["Gender"].apply(lambda x: clean_gender(x))
print(features.head())
```

Для кодирования номинальных данных используется метод one-hot-encoding. Создается столбец для каждого значения и используется 1 и 0 для обозначения выражения каждого значения. Эти новые столбцы часто называются фиктивными переменными.

Метод One-Hot Encoding (одноразрядное кодирование) используется для преобразования номинальных (категориальных) данных в числовую форму, которую можно использовать для обучения нейронных сетей.

Суть метода заключается в следующем:

1️⃣ Каждая уникальная категория в номинальном признаке представляется в виде отдельного бинарного признака (столбца). То есть, если у вас есть признак "Цвет" с категориями "Красный", "Зеленый" и "Синий", то будет создано три новых бинарных признака.

2️⃣ В каждом бинарном признаке значение 1 указывает, что объект принадлежит к соответствующей категории, а значение 0 — что объект не принадлежит к этой категории.

3️⃣ Другие числовые признаки в данных могут быть оставлены без изменений или также преобразованы в числовую форму при необходимости.

Преимущества One-Hot Encoding:

✅ Позволяет представить номинальные данные в удобной числовой форме без введения искажений или порядка между категориями.

✅ Подходит для использования с нейронными сетями, которые ожидают числовые входные данные.

Но есть и некоторые соображения, которые стоит учитывать:

❗️ One-Hot Encoding может привести к увеличению размерности данных при большом количестве уникальных категорий. Это может привести к проблеме известной как "проклятие размерности".

❗️ В случае больших данных и множества категорий, использование One-Hot Encoding может быть вычислительно затратным.

Важно правильно применять One-Hot Encoding в соответствии с контекстом и требованиями вашей конкретной задачи и данных.

### Нормализация данных

Алгоритмы машинного обучения, как правило, работают лучше или сходятся быстрее, когда различные функции (переменные) имеют меньший масштаб. Поэтому перед обучением на них моделей машинного обучения данные обычно нормализуются.

Нормализация также делает процесс обучения менее чувствительным к масштабу функций. Это приводит к улучшению коэффициентов после тренировки.

**Нормализация на стандартное отклонение (StandartScaler)** - используется для изменения размера распределения значений так, чтобы среднее значение наблюдаемых значений было равно 0, а стандартное отклонение – 1.

**Нормализация Min-max** устанавливает наименьшее наблюдаемое значение равным 0, а наибольшее наблюдаемое значение — 1.


```python
#нормализация на стандартное отклонение
from sklearn.preprocessing import StandardScaler
scale_features_std = StandardScaler()
features[['Age']]=scale_features_std.fit_transform(features[['Age']])
print(features.head())
```

### Разделение данных для обучения и тестирования
Для обучения модели данные обычно разделяют на три основные выборки: обучающую выборку (training set), проверочную выборку (validation set) и тестовую выборку (test set).

1️⃣ Обучающая выборка (Training Set):
   Обучающая выборка используется для обучения нейронной сети. Она содержит набор образцов данных, на которых сеть будет тренироваться. Обычно это большая часть доступных данных, например, 70-80% от общего объема.

2️⃣ Проверочная выборка (Validation Set):
   Проверочная выборка используется для настройки гиперпараметров модели, таких как количество слоев, количество нейронов и скорость обучения. Эта выборка помогает оценить производительность модели на данных, которые она ранее не видела. Обычно это около 10-20% от общего объема данных.

3️⃣ Тестовая выборка (Test Set):
   Тестовая выборка используется для окончательной оценки производительности модели после ее обучения и настройки. Эта выборка должна быть независимой от обучающей и проверочной выборок. Она помогает оценить обобщающую способность модели на новых, непредставленных ей данных. Обычно это около 10-20% от общего объема данных.

🔹 Важно сохранять следующие принципы при разделении данных:

- Разделение должно быть случайным, чтобы избежать смещения или корреляции между выборками.
- Данные в каждой выборке должны быть представительными для общего распределения данных.

Применение хорошей стратегии разделения данных помогает оценить и сравнить производительность модели с внешними данными и предотвращает переобучение (overfitting), когда модель показывает хорошую производительность только на обучающих данных, но плохо справляется с новыми данными.
