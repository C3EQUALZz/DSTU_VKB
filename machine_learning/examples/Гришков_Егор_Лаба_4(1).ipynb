# Задание 1

    Создайте Series из последовательности 15 значений, равномерно разбивающих отрезок [0, 20] (воспользуйтесь функцией linspace)
    Определите отношение элементов полученной серии к их предыдущим элементам.
    В результате необходимо получить среднее полученного вектора, оставив в нём только те значения, которые не более чем 1.5.

Пояснения:

    Если было бы необходимо найти последовательность из 3-х значений,
    равномерно разбивающих отрезок [0,1], то это были бы значения [0, 0.5, 1].

    Если был бы дан список элементов a = [1,2,3,12], отношения элементов к предыдущим будут равны [NaN, 2, 1.5, 4]. А на последнем этапе в таком примере останется только [1.5] и среднее значение будет также 1.5.


```python
import pandas as pd
import numpy as np

s = pd.Series(np.linspace(0, 20, 15))
print(s)
```

```python
import pandas as pd
import numpy as np

s = pd.Series(np.linspace(0, 20, 15))
print(s)

s = (s/s.shift(1)).where(s/s.shift(1) <= 1.5, s/s.shift(1))
print(s)
```

```python
import pandas as pd
import numpy as np

s = pd.Series(np.linspace(0, 20, 15))
print(s)

s = (s/s.shift(1)).where(s/s.shift(1) <= 1.5)
print(s)
print(s.mean())
```

```python
import pandas as pd
import numpy as np

np.random.seed(242)
s = pd.Series(data=np.random.normal(size=100))
print(s)

s = s ** 3
s.index *= 3
print(s)

sum1 = s.where((s < 2.6) & (s.index % 2 != 0)).sum()
print(sum1)
sum_ = s[(x % 2 != 0 for x in s.index) and (s < 2.6)].sum()
print(sum_)
```

```python
import pandas as pd
import numpy as np

np.random.seed(242)
s = pd.Series(data=np.random.normal(size=100))
print(s)

s = s ** 3
s.index *= 3
print(s)

sum1 = s.where((s.index % 2 != 0) & (s < 2.6)).sum()
print(sum1)
sum_ = s[(x % 2 != 0 for x in s.index) and (s < 2.6)].sum()
print(sum_)

count1 = s.where(s < 0).count()
count_ = s[s < 0].count()
print(count1)
print(count_)
```

# Задание 8
Определите модуль разницы между средними тратами женщин и мужчин (трата - отрицательное значение amount). (*)

Выведите ответ в виде вещественного числа, округлённого до двух знаков после запятой, отделив дробную часть точкой в формате "123.45"
Пояснения:

(*) Если в результате для мужчин получились значения [-1,-3,-5], а для женщин [-1,-2,-3], то модуль разницы между средними арифметическими -3 и -2 будет равен 1.

```python
import pandas as pd

transactions = pd.read_csv(filepath_or_buffer=r'/transactions.csv', sep=',')
tr_mcc_codes = pd.read_csv(filepath_or_buffer=r'/tr_mcc_codes.csv', sep=';')
tr_types = pd.read_csv(filepath_or_buffer=r'/tr_types.csv', sep=';')
gender_train = pd.read_csv(filepath_or_buffer=r'/gender_train.csv', sep=',')
# print(gender_train)
# print(tr_mcc_codes)
# print(tr_types)
# print(transactions)

df = pd.merge(transactions, gender_train, how='left')


df1 = df[df['gender'] == 1.0]

df1 = df1[df1['amount'] < 0]
df2 = df1['amount'].sum()

df1_1 = df[df['gender'] == 0.0]

df1_1 = df1_1[df1_1['amount'] < 0]
df2_1 = df1_1['amount'].sum()


print(round(abs(df2_1 - df2), 2))
```
